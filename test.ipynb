```python
import pandas as pd
import os

file_path_09_10 = 'data/online_retail_09_10.csv'
file_path_10_11 = 'data/online_retail_10_11.csv'
```

```python

df_09_10 = pd.read_csv(file_path_09_10, encoding='ISO-8859-1')
df_10_11 = pd.read_csv(file_path_10_11, encoding='ISO-8859-1')

print(f"Fichier 09-10 : {df_09_10.shape}")
print(f"Fichier 10-11 : {df_10_11.shape}")
```

```python
# 2. Fusionnement des deux DataFrames
df_merged = pd.concat([df_09_10, df_10_11], ignore_index=True)

print(f"DataFrame fusionné : {df_merged.shape}")
df_merged.head()
```

```python
df_products = df_merged[['StockCode', 'Description', 'UnitPrice']].drop_duplicates()
df_products.to_csv('data/products.csv', index=False)

df_orders = df_merged[['ï»¿InvoiceNo', 'InvoiceDate', 'CustomerID', 'Country']].drop_duplicates()
df_orders.to_csv('data/orders.csv', index=False)

df_orders_items = df_merged[['ï»¿InvoiceNo', 'StockCode', 'Quantity']]
df_orders_items.to_csv('data/orders_items.csv', index=False)

print("Les fichiers products.csv, orders.csv et orders_items.csv ont été créés avec succès.")
```
